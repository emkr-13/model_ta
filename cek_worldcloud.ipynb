```python
import pandas as pd
```

```python
df = pd.read_csv('online_news_50000_clean_label_all_baru.csv')
df.head(10)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
sentiments = df['sentimen'].unique()

for sentiment in sentiments:
    text = " ".join(content for content in df[df['sentimen'] == sentiment]['content_clean'])
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {sentiment.capitalize()} Sentiment')
    plt.show()
    
    time.sleep(2)  # Pause for 2 seconds before showing the next word cloud
```

```python
import pandas as pd
from concurrent.futures import ThreadPoolExecutor

def filter_data(chunk):
    # Membuat filter kata-kata yang diinginkan
    keywords = ['pemilu', 'mk', 'jokowi', 'kpu', 'ganjar', 'prabowo', 'anies', 'cak imin', 'gibran', 'mahfud md', 'capres', 'cawpres']
    
    # Membuang baris dengan nilai yang hilang (NaN) dalam kolom 'processed_text'
    chunk = chunk.dropna(subset=['processed_text'])
    
    # Membuat filter dengan menggunakan fungsi .contains()
    return chunk[chunk['processed_text'].str.contains('|'.join(keywords), case=False)]

# Membaca file CSV
df = pd.read_csv('pre_content_new_all.csv')

# Menentukan jumlah thread yang akan digunakan
num_threads = 4

# Membagi data menjadi beberapa chunk untuk diproses oleh thread yang berbeda
chunk_size = len(df) // num_threads

# Inisialisasi ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=num_threads) as executor:
    # Memproses setiap chunk data secara paralel
    results = list(executor.map(filter_data, [df.iloc[i:i+chunk_size] for i in range(0, len(df), chunk_size)]))

# Menggabungkan hasil pemrosesan dari setiap thread
filtered_df = pd.concat(results)

# Menyimpan hasil filter ke dalam file CSV baru
filtered_df.to_csv('filtered_content_baru.csv', index=False)

# Menampilkan hasil filter
filtered_df
```
